```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

ticker = ['^GSPC','^VVIX','^VIX']
data = yf.download(ticker, '2020-01-01')['Close']

data['VVIX/VIX'] = data['^VVIX']/data['^VIX']

data = data.dropna()

plt.style.use('dark_background')
fig, (ax) = plt.subplots(figsize=(16,8))
data['^GSPC'].plot(ax=ax,title="SP500 y VVIX/VIX")
ax = plt.gca().twinx()
data['VVIX/VIX'].plot(ax=ax, color="r")
plt.show()
```
